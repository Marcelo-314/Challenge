**Question 1:** How many customers are in the file?
(as help, we have added some comments and starter code to help you structure the solution)

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")


with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')

  next(csvreader)

  count = 0
  for row in csvreader:
      count += 1

print("Number of customers:", count)


print("All done!")
```

**Question 2:** In how many different states do the customers live in?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

states = set()

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')

    header = next(csvreader)
    state_idx = header.index("State")

    for row in csvreader:
        state = row[state_idx].strip().upper()
        states.add(state)

print("Number of different states:", len(states))
print("States:", states)
```

**Question 3** What is the state with most customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

state_counts = {}

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')

    header = next(csvreader)
    state_idx = header.index("State")

    for row in csvreader:
        state = row[state_idx].strip().upper()   # normalizamos
        state_counts[state] = state_counts.get(state, 0) + 1

max_state = None
max_count = 0
for st, cnt in state_counts.items():
    if cnt > max_count:
        max_state = st
        max_count = cnt

print("State with most customers:", max_state, "with", max_count, "customers")
```

**Question 4** What is the state with the least customers?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

state_counts = {}

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')

    header = next(csvreader)
    state_idx = header.index("State")

    for row in csvreader:
        state = row[state_idx].strip().upper()
        state_counts[state] = state_counts.get(state, 0) + 1

min_count = min(state_counts.values())
least_states = [st for st, cnt in state_counts.items() if cnt == min_count]

print("States with the least customers:", least_states, "with", min_count, "customer each")
```

**Question 5:** What is the most common last name?

```python
from os.path import exists
import csv

datafile = "./sample_data/customers.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")

last_name_counts = {}

with open(datafile, 'r') as f:
    csvreader = csv.reader(f, delimiter=',')

    header = next(csvreader)
    last_idx = header.index("LastName")

    for row in csvreader:
        last = row[last_idx].strip()
        last_name_counts[last] = last_name_counts.get(last, 0) + 1

most_common_last = max(last_name_counts, key=last_name_counts.get)
print(
    "Most common last name:",
    most_common_last,
    "appears",
    last_name_counts[most_common_last],
    "times",
)
```

**Question #1:** How many unique orders are in the orders.csv file?

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

**Question #3:** What is the highest number of items per order?

**Question #4:** What is the number of orders placed in October 2021?

**Question #5:** Which customer spent the most amount of money in 2021?

**Question #6:** Historically, what is the best month for sales?

Once you get your answers, remember to go back to the course and introduce them in the multiple choice quiz

```python
from os.path import exists
import csv
from collections import defaultdict
from datetime import datetime
import calendar

datafile = "./sample_data/orders.csv"
if not exists(datafile):
    raise SystemExit("You should run the first code cell and download the dataset files!")


items_per_order = defaultdict(int)
order_date      = {}
order_total     = {}
order_customer  = {}

with open(datafile, newline='') as f:
    reader = csv.DictReader(f)

    for row in reader:
        oid  = row["OrderID"].strip()
        cid  = row["CustomerID"].strip()
        dstr = row["Date"].strip()
        tstr = row["OrderTotal"].strip()

        items_per_order[oid] += 1

        order_customer[oid] = cid

        if oid not in order_total:
            order_total[oid] = float(tstr)

            if dstr:
                order_date[oid] = datetime.fromisoformat(dstr)
            else:
                order_date[oid] = None


num_unique_orders = len(items_per_order)
print("Unique orders:", num_unique_orders)

total_items = sum(items_per_order.values())
avg_items_per_order = round(total_items / num_unique_orders, 2)
print("Average items per order:", avg_items_per_order)

max_items = max(items_per_order.values())
print("Max items in a single order:", max_items)

orders_oct_2021 = {
    oid for oid, d in order_date.items()
    if d is not None and d.year == 2021 and d.month == 10
}
print("Orders in October 2021:", len(orders_oct_2021))

spent_2021 = defaultdict(float)
for oid, d in order_date.items():
    if d is not None and d.year == 2021:
        cid = order_customer[oid]
        spent_2021[cid] += order_total[oid]

top_customer_id = max(spent_2021, key=spent_2021.get)
print("Top customer ID in 2021:", top_customer_id)
print("Amount spent:", spent_2021[top_customer_id])


sales_by_month = defaultdict(float)

for oid, dt in order_date.items():
    if dt is None:
        continue
    month = dt.month
    sales_by_month[month] += order_total[oid]

best_month_num = max(sales_by_month, key=sales_by_month.get)
best_month_name = calendar.month_name[best_month_num]

print("Best month number:", best_month_num)
print("Best month name:", best_month_name)
print("Total sales in that month:", sales_by_month[best_month_num])



```
